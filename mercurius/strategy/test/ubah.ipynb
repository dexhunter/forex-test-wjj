# Uniform Buy and Hold (ubah.m)

```python
import numpy as np

def ubah(fid, data, *args, opts):
    '''
    `This file is an entry for the market strategy.

 function [cumRet, cumProdRet, dailyRet, dailyPortfolio] ...
     = ubah(fid, data, varargins, opts)
 cumRet: cumulative wealth achived at the end of a period.
 cumProdRet: cumulative wealth achieved till the end each period.
 dailyRet: daily return achieved by a strategy.
 dailyPortfolio: daily portfolios

 data: market sequence vectors
 fid: handle for write log file
 varargins: variable parameters
 opts: option parameter for behvaioral control

 Example: [cumRet, cumProdRet, dailyRet, dailyPortfolio] ...
          = market_start(fid, data, {0}, opts);

    
    '''
    tc = args[1]
    return tc

def ubah_kernel(data, weight_o):
    return ubah_expert(data, weight_o)

def ubah_expert(data, weight_o):
    '''simply output last portfolio, for buy and hold market'''
    return weight_o
    
    
    
def ubah_run(fid, data, tc):
    # first get tc
    
    
    n, m = data.shape #n:rows, m:cols #n:507 m:30
    
    #print('n',n)
    #print('m',m)
    #print(data[:,1].shape)
    # variables for return, start with uniform weight
    cum_ret = 1
    cumprod_ret = np.ones((n, 1), float) #507,1
    # print('shape of cumpro_ret is', cumprod_ret.shape) 
    
    daily_ret = np.ones((n, 1), float) #507,1
    
    day_weight = np.ones((m, 1), float)/m  #30,1
    #  print('shape of day_weight is', day_weight.shape)
    day_weight_o = np.zeros((m, 1), float) # 30,1
    daily_portfolio = np.zeros((n,m)) #507,30
    
    #print log file head
    print(fid, '-------------------------------------\n')
    print(fid, 'Parameters [tc: %f].\n' % tc)
    print(fid, 'day\t Daily Return\t Total return\n')
    
    print(1, '-------------------------------------\n')
    
    '''
      if opts.quiet_mode == False:
        print(1, 'Parameters [tc: %f].\n' % tc)
        print(1, 'day\t Daily Return\t Total return\n')
          
    '''

    #backtest
    #TODO: double check if we need to use np.arange(1, n+1)
    for t in range(n):
        
        # Calculate t's portfolio at the beginning of t-th trading day
        if t >= 1:
            ''' ~~TODO: What is [] in matlab?~~
                It's array in matlab
            '''
            day_weight = np.array(ubah_kernel(data[0:t-2,:], day_weight_o))
            
        # Normalize the constraint, always useless??
        
        day_weight = np.divide(day_weight, np.sum(day_weight, axis=0)) # elementwise divide and column-sum
        #print('day_weight shape', day_weight.shape)
        daily_portfolio[t,:] = day_weight.conj().T
        
        #print('daily_portfolio shape',daily_portfolio.shape)
        
        # Cal t's return and total return
        daily_ret[t, 0] = np.dot(data[t,:],day_weight)*(1-tc/2*np.sum(np.absolute(day_weight-day_weight_o) ))
                
        cum_ret = cum_ret * daily_ret[t,0]
        cumprod_ret[t, 0] = cum_ret
        
        #Adjust weight[t, :] for the transaction cost issue
        #print('data shape', data[t,:].shape)
        day_weight_o = day_weight * data[t,:][:,None] / daily_ret[t, 0]
        #print('day_weight_o shape',day_weight_o.shape)

        print(fid, '%d\t%f\t%f\n' % (t, daily_ret[t,0], cumprod_ret[t,0]))
        
              
        #TODO: need to double check if np.mod works
        '''
        if opts.quiet_mode == False:
            if np.mod(t, opts.display_interval) == 0:
                  print(1, '%d\t%f\t%f\n' % (t, daily_ret[:t,:1], cumprod_ret[:t,:1]))     
        '''
            
        
    print(fid, 'Market(tc=%.4f), Cumulative return: %.2f\n'% (tc, cum_ret))
    print(fid, '-------------------------------------\n')
    print(1, 'Market(tc=%.4f), Cumulative return: %.2f\n'% (tc, cum_ret))
    print(1, '-------------------------------------\n')
              
              
import scipy.io as sio
djia = sio.loadmat('djia.mat') # load a test input
print(djia['data'].shape)
ubah_run(1, djia['data'], 0.5 )

'''

The original algorithm in matlab has final result

507     1.006276        0.764361


and initial result


'''
```
